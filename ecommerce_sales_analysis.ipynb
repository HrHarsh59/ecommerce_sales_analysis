```python
import seaborn as sns
plt.figure(figsize=(8,5))
sns.countplot(df, x=df["category"])
plt.show()
```

```python
df["total_sales"] = df[[f'sales_month_{i}' for i in range(1,13)]].sum(axis=1)
```

```python
df.head(100)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=df["category"], y=df["total_sales"])
plt.show()
```

```python
category_sales = df.groupby("category")["total_sales"].sum().reset_index().sort_values(by="total_sales", ascending= False)
```

```python
category_sales.head()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="category", y = "total_sales", data=category_sales, palette="viridis")
plt.title("Top Perfoming Cetegory")
plt.ylabel("Total Sales")
plt.xlabel("Category")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="price", y = "total_sales", data=df, hue="category", palette = "deep", alpha= 0.7)
plt.title("Price vs Sales")
plt.xlabel("Price")
plt.ylabel("Total Sales")
plt.legend(loc="upper right", bbox_to_anchor=(1.25,1))
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="price", y = "review_score", data=df, hue="category", palette = "deep", alpha= 0.7)
plt.title("Price vs Customer Review")
plt.xlabel("Price")
plt.ylabel("Customer Review")
plt.legend(loc="upper right", bbox_to_anchor=(1.25,1))
plt.show()
```

```python
plt.figure(figsize=(10,5))
corr = df[["price", "review_score", "total_sales"]].corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
monthly_sales = df[[f'sales_month_{i}' for i in range(1,13)]].sum().reset_index()
monthly_sales.columns = ["month", "total_sales"]
monthly_sales["month"] = monthly_sales["month"].apply(lambda x: int(x.split('_')[-1]))
monthly_sales
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(x=monthly_sales["month"], y=monthly_sales["total_sales"], markers="o")
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(range(1,13))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='review_count', y='review_score', data=df, size='review_count', hue='review_score', palette='viridis', sizes=(20, 200))
plt.title('Review Score vs. Review Count')
plt.xlabel('Review Count')
plt.ylabel('Review Score')
plt.legend(title='Review Analys', loc="upper right", bbox_to_anchor=(1.25,1))
plt.show()
```
